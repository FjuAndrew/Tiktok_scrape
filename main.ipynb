```python
from DrissionPage import ChromiumPage, ChromiumOptions
import  pandas as pd
import time
page = ChromiumPage(2222)
page.get('https://www.tiktok.com/@geevideo') #輸入自己想要抓取的頻道首頁網址

def get_basic_info():
    #關注中 title=關注中
    ele = page.ele('@title=關注中')
    sees = ele.text
    #粉絲 title=粉絲
    ele1 = page.ele('@title=粉絲')
    subs = ele1.text
    #讚數 title=讚
    ele2 = page.ele('@title=讚')
    likes = ele2.text
    print("關注中:",sees+"\n粉絲數:",subs+"\n讚數:",likes)


def get_content():
    data =[]
    #影片標題 
    titles = page.eles('.css-vi46v1-DivDesContainer eih2qak4')
    #影片觀看數
    video_views = page.eles('.video-count css-dirst9-StrongVideoCount e148ts222')
    for i in range(len(titles)):
        if i > 99:
            print('已達100支影片上限')
            break
        title = titles[i].text
        view = video_views[i].text
        print(f'No.{i+1}影片標題: {title}')
        print('觀看數:',view)
        data.append([title, view])
    df = pd.DataFrame(data,columns=['影片標題','觀看數'])
    return df

def login():
    try:
        #如出現登入畫面
        ele = page.ele('.css-1gtmaw0-DivBoxContainer e1cgu1qo0')
        ele.click()
    except Exception as e:
        print("無須登入")


def load_more(scroll_times):#往下滑載入更多影片
    for i in range(scroll_times):
            print(f'滾第{i+1}次')
            titles = page.eles('.css-vi46v1-DivDesContainer eih2qak4')    
            page.scroll.to_see(titles[-1])
            time.sleep(5)
login()
get_basic_info()

load_more(3)
df =get_content()
print(f'抓到影片標題隻數:{len(df)}')
#print(df)
```
